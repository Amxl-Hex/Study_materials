```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/cup.jpeg")
cv2_imshow(img)
```

```python
import numpy as np
print(type(img))
print(img.shape)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
print(img_gray.shape)
```

```python
cv2.imwrite("cup_bw.jpeg", img_gray)
```

```python
img_gray_resize = cv2.resize(img_gray, (200, 200))
cv2_imshow(img_gray_resize)
print(img_gray_resize.shape)
```

```python
img_resize = cv2.resize(img, (200, 200))
cv2_imshow(img_resize)
print(img_resize.shape)
```

```python
img_clrb = img[:, :, 0] #blue
img_clrg = img[:, :, 1] #green
img_clrr = img[:, :, 2] #red
new_img = np.hstack((img_clrb, img_clrg, img_clrr))
cv2_imshow(new_img)
print(new_img.shape)
```

```python
img_flipbt = cv2.flip(img, 0) #bottom to top
img_fliprl = cv2.flip(img, 1) #left to right
img_flipboth = cv2.flip(img, -1) #both
img_flip = np.hstack((img_flipbt, img_fliprl, img_flipboth))
cv2_imshow(img_flip)
```

```python
img_crop = img[75:250, 25:150]
cv2_imshow(img_crop)
```

```python
img = cv2.imread("/content/cup.jpeg")
img_rectangle = img.copy()
cv2.rectangle(img_rectangle, pt1 = (45, 50), pt2 = (210, 180), color = (0, 255, 0), thickness = 3)
cv2_imshow(img_rectangle)
```

```python
img_circle = img.copy()
cv2.circle(img_circle, center = (120, 105), radius = 90, color = (0, 0, 255), thickness = 3)
cv2_imshow(img_circle)
```

```python
cv2_imshow(img)
```

```python
img_line = img.copy()
cv2.line(img_line, pt1 = (10, 50), pt2 = (210, 50), color = (190, 50, 50), thickness = 3)
img_text = img_line.copy()
cv2.putText(img_text, text = 'Amal Cherian Jacob', org = (10,40), fontFace = cv2.FONT_HERSHEY_SCRIPT_COMPLEX, fontScale = 0.7, color = (0, 0, 0), thickness = 1)
img_linetext = np.hstack((img_line, img_text))
cv2_imshow(img_linetext)
```

```python
img = cv2.imread("/content/cup.jpeg")
img_cline = img.copy()
attr = img_cline.shape
print(attr)
print(attr[0])
print(attr[1])
widthhalf = int(attr[1]/2)
height = int(attr[0])
cv2.line(img_cline, pt1 = (widthhalf, 0), pt2 = (widthhalf, height), color = (200, 100, 250), thickness = 2)
cv2_imshow(img_cline)
```
