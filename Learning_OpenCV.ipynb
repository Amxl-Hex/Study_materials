```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/noise_picture.jpeg", cv2.IMREAD_GRAYSCALE)
gaussian_img1 = cv2.GaussianBlur(img, ksize = (5, 5), sigmaX = 0)
gaussian_img2 = cv2.GaussianBlur(img, ksize = (5, 5), sigmaX = .7)
gaussian_img3 = cv2.GaussianBlur(img, ksize = (5, 5), sigmaX = .9)
img_gaussian = np.hstack((img, gaussian_img1, gaussian_img2, gaussian_img3))
cv2_imshow(img_gaussian)
```

```python
bilateral_img1 = cv2.bilateralFilter(img, d = 5, sigmaColor = 75, sigmaSpace = 75)
bilateral_img2 = cv2.bilateralFilter(img, d = 9, sigmaColor = 100, sigmaSpace = 200)
bilateral_img3 = cv2.bilateralFilter(img, d = 9, sigmaColor = 50, sigmaSpace = 150)
img_bilateral = np.hstack((img, bilateral_img1, bilateral_img2, bilateral_img3))
cv2_imshow(img_bilateral)
```

```python
median_img1 = cv2.medianBlur(img, ksize = 5)
median_img2 = cv2.medianBlur(img, ksize = 9)
median_img3 = cv2.medianBlur(img, ksize = 15)
img_median = np.hstack((img, median_img1, median_img2, median_img3))
cv2_imshow(img_median)
```

```python
mean_img1 = cv2.blur(img, ksize = (5,5))
mean_img2 = cv2.blur(img, ksize = (9,5))
mean_img3 = cv2.blur(img, ksize = (5,9))
img_mean = np.hstack((img, mean_img1, mean_img2, mean_img3))
cv2_imshow(img_mean)
```

```python
img = cv2.imread('/content/flower.jpeg', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5,5), np.uint8)
dilation1 = cv2.dilate(img, kernel, iterations = 1)
dilation2 = cv2.dilate(img, kernel, iterations = 2)
dilation3 = cv2.dilate(img, kernel, iterations = 3)
img_dilation = np.hstack((img, dilation1, dilation2, dilation3))
cv2_imshow(img_dilation)
```

```python
eroded1 = cv2.erode(img, kernel, iterations = 1)
eroded2 = cv2.erode(img, kernel, iterations = 2)
eroded3 = cv2.erode(img, kernel, iterations = 3)
img_eroded = np.hstack((img, eroded1, eroded2, eroded3))
cv2_imshow(img_eroded)
```

```python
img_allfilters = np.vstack((img_gaussian, img_bilateral, img_median, img_mean, img_dilation, img_eroded))
cv2_imshow(img_allfilters)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/noise_picture.jpeg", cv2.IMREAD_GRAYSCALE)
ret, otsu_thresh1 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
ret, otsu_thresh2 = cv2.threshold(img, 127, 127, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
otsu_img = np.hstack((img, otsu_thresh1, otsu_thresh2))
cv2_imshow(otsu_img)
```

```python
adaptive_gaussian = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
adaptive_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
img_adaptive = np.hstack((img, adaptive_gaussian, adaptive_mean))
cv2_imshow(img_adaptive)
```
